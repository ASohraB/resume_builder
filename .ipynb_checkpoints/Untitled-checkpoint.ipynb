```python
import requests
import re

from bs4 import BeautifulSoup


publication_url = "https://medium.com/data-science-in-your-pocket"
profile_url = 'https://medium.com/@mehulgupta_7991/about'
page = requests.get(profile_url)
soup = BeautifulSoup(page.content, 'html.parser')
followers = re.findall('(\d+) Followers',page.text)[0]
following = re.findall('(\d+) Following',page.text)[0]
print(followers,following)
```
